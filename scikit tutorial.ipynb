# KNN with Iris

Thne K Nearest neighbors algorithm is contained within scikit and is used for taking a dataset and assigning input and outputs from the dataset that are in line with what you want to do with the dataset. After assigning the inputs and outputs, you can then train the datset with 1 command and the values from each of the inputs are connected in the algorithm with their relevant output. Once that is done, with another 1 command, you can make predictions based on the connections made in the algorithm between the inputs and outputs. Here we will use the iris dataset to illustrate

## Libraries

```python
import sklearn.neighbors as nei
import pandas as pd
```

## Data

```python
# load the irid datset from a url
df = pd.read_csv("https://github.com/ianmcloughlin/datasets/raw/master/iris.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
# load the seaborn package
import seaborn as sns

# Plot the iris dataset with pair plot
x = sns.pairplot(df, hue="class")
```

## Inputs and outputs

```python
# Use the KNN algorithm to use the inputs below to predict the class of iris it is
# Declare the inputs and the outputs for the knn algorithm
inputs = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
outputs = df['class']
```

## Classifier

```python
# Create an object/data structure that represents the k nearest neighbors algorithm, where k is set to 5. 
knn = nei.KNeighborsClassifier(n_neighbors=5)
# This will pick the 5 nearest neighbours from the dataset
```

## Fit

```python
# Train the knn algorithm on the specific inputs and outputs defined above, with the classifier knn
knn.fit(inputs, outputs)
# Now, knn has the classifier trained with the specific inputs and outputs 
```

## Predict

### Use figures that are different than the input values and output values that were used to train the predictor

Instead of using your full dataset to train the algorithm, use 80% of the dataset to train and 20% to test afterwards

```python
# Use the train-test split method to do this
import sklearn.model_selection as mod
```

```python
mod.train_test_split(inputs, outputs, test_size=0.33)
# mod.train_test_split(inputs, outputs, test_size=0.33)  3 parameters: inputs, outputs, test_size=0.33)
# inputs is indicating where the inputs are taken from
# outputs is indicating where the uptputs are taken from
# test_size=0.33 is indicating to retain 1/3 of thr dataset from the train and use in test
# This gives some sort of output, but it is difficult to make sense of it. See the next command to better expand and explain
```

```python
# Below, because the test_size is a parameter, this is effectively creating 2 datasets from the original one: 1 for train
# and 1 for test. So for the input parameter we have 2 types of input: input_train and input_test. Similarly for the 
# outputs parameter we have output_train and output_test
inputs_train, inputs_test, outputs_train, outputs_test = mod.train_test_split(inputs, outputs, test_size=0.33)
```

```python
# So, using train_test_split method, we can do the following
# 
inputs_train, inputs_test, outputs_train, outputs_test = mod.train_test_split(inputs, outputs, test_size=0.33)
knn = nei.KNeighborsClassifier(n_neighbors=5)
knn.fit(inputs_train, outputs_train)
```

```python
(knn.predict(inputs_test) == outputs_test).sum()
```
